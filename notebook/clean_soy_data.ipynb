```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('../data/soybean.csv', header=None)
print(data.shape)
data.head()
```

```python
# Get unique values of each column and if there is only one unique value, drop the column
for col in data.columns:
    if len(data[col].unique()) == 1:
        data.drop(col, axis=1, inplace=True)

data.head()
```

```python
# Transform the last column to numerical values
data.iloc[:, -1] = LabelEncoder().fit_transform(data.iloc[:, -1])
data.head()
```

```python
# Save the dataframe except for the last column to a npy file
data_without_target = data.iloc[:, :-1]
data_without_target = data_without_target.to_numpy()
np.save('../data/soy_data.npy', data_without_target)

# Save the last column to a npy file
# The last column is the cluster number
target = data.iloc[:, -1].to_numpy()
np.save('../data/soy_target.npy', target)
```
